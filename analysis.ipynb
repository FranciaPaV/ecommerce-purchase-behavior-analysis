### [C4] Para cada cliente, ¿qué proporción de sus productos ya los había pedido?

```python
# Unir las tablas 'orders' y 'order_products' para obtener 'user_id' en cada pedido
user_order_products = pd.merge(
    orders[['order_id', 'user_id']],
    order_products[['order_id', 'product_id', 'reordered']],
    on='order_id',
    how='inner'
)

# Calcular métricas por cliente
user_reorder_stats = (
    user_order_products.groupby('user_id')
    .agg(
        total_products=('reordered', 'count'),      # Total de productos pedidos
        reordered_products=('reordered', 'sum')     # Suma de reórdenes (reordered=1)
    )
    .reset_index()
)

# Calcular proporción y redondear
user_reorder_stats['reorder_proportion'] = (
    (user_reorder_stats['reordered_products'] / user_reorder_stats['total_products']) * 100
).round(2)

# Mostrar resultados ordenados (ejemplo: top 10 clientes con mayor proporción)
print(user_reorder_stats.sort_values('reorder_proportion', ascending=False).head(10))
```

Clientes con 100% de reórdenes y alto volumen son candidatos ideales para programas B2B.

Casos con bajo volumen requieren validación de datos.
